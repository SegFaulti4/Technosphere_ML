```python
import numpy as np
import pandas as pd
import folium
import matplotlib.colors as colors
import matplotlib.cm as cm
```

```python
df = pd.read_csv("Rus_schools_final.csv", encoding='1251')

# добавил свою школу в базу, удалил все школы, не расположенные в Москве

mySchool = {
    'Unnamed: 0': df.size + 100000,
    'lat': 55.694718,
    'lon': 37.524362,
    'name': 'Университетская гимназия МГУ имени М.В.Ломоносова',
    'struct': '(Государственное образовательное учреждение)',
    'addr': '119192, Москва, Ломоносовский проспект 27, корпус 7'
}
df = pd.concat([df, pd.DataFrame([mySchool])], ignore_index=True)
df = df[~(df['lat'].isnull() | df['lon'].isnull())]
df = df[df['addr'].str.contains('Москва,')]
```

```python
# центрировал карту относительно своей школы

map_schools = folium.Map(location=(mySchool['lat'], mySchool['lon']),
                            tiles='Stamen Toner', 
                            zoom_start=13)

norm = colors.Normalize(vmin=0, vmax=50000)
cmap = cm.plasma
colormap= cm.ScalarMappable(norm=norm, cmap=cmap)

# jupyter в комбинации с folium отображает русский текст некорректно

for ind, row in df.iterrows():
    folium.CircleMarker([row['lat'], row['lon']],
                        radius=7,
                        popup='{}'.format('<br>'.join(map(str, [row['name']]))),
                        color=colors.rgb2hex((colormap.to_rgba(row['Unnamed: 0']))),
                        fill_color=colors.rgb2hex((colormap.to_rgba(row['Unnamed: 0']))),
                        fill=True, 
                        fill_opacity=0.9
                       ).add_to(map_schools)
    
map_schools
```
